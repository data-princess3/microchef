```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/adminsha/Downloads/applications(2.0).csv')
df1 = pd.read_csv('/Users/adminsha/Downloads/industries(2.0).csv')
df.head(12)
```

```python
df.info()
```

```python
without_dub = df.drop_duplicates(subset=['applicant_id'], inplace=True)
print(without_dub)
```

```python
df['External Rating'] = df['External Rating'].fillna(0)
df['External Rating']
```

```python
df['Education level'] = df['Education level'].fillna("Середнє")
df['Education level']
```

```python
df1.head(12)
```

```python
df2 = pd.merge(df, df1)
```

```python
df2['AgeCriterium'] = df2['Age'].apply(lambda x: 20 if 35 < x < 55 else 0)
```

```python
df2['MarriageCriterium'] = df2['Marital status'].apply(lambda x: 20 if x == 'Married' else 0)
```

```python
df2['LocationCriterium'] = df2['Location'].apply(lambda x: 10 if x == 'Київ чи область' else 0)
```

```python
df2['ExtCriterium'] = df2['External Rating'].apply(lambda x: 20 if x >= 7 else (-20 if x < 2 else 0))
df2.head(7)
```

```python
df['Applied at'] = pd.to_datetime(df['Applied at'], format = 'mixed')
df.head(7)
```

```python
df.info()
```

```python
df['Applied at'].dt.day_name()
```

```python
df2['Day']=df['Applied at'].dt.day_name()
df2.head()
```

```python
df2['DayCriterium'] = df2['Day'].apply(lambda x: 20 if x in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] else 0)
df2.head(7)
```

```python
df2['Total_score'] = df2['Score'] + df2['AgeCriterium'] + df2['MarriageCriterium'] + df2['LocationCriterium']+ df2['ExtCriterium']+ df2['DayCriterium']
```

```python
df2['Total_score'] = df2['Total_score'].apply(lambda x: max(0, min(x, 100)))
```

```python
df2['Total_score'] = df2.apply(lambda row: 0 if pd.isna(row['Amount']) or row['External Rating'] == 0 else max(0, min(row['Total_score'], 100)), axis=1)
```

```python
df2 = df2[df2['Total_score'] != 0]
df2.head()
```

```python
df3 = pd.DataFrame(df2)
```

```python
df3['Applied at'] = pd.to_datetime(df['Applied at'], format='mixed')
```

```python
df3['submission_week'] = df3['Applied at'].dt.isocalendar().week
```

```python
pivot_table = df3.pivot_table(values='Total_score', index='submission_week', aggfunc='mean')
print(pivot_table)
```
